```python
import pandas as pd
```

```python
df = pd.read_csv('air_final_clean.csv')
```

```python
df.head()
```

```python
df.room_type = df.room_type.map({'Entire home/apt': 'Entire Home/Apt', 'Private room': 'Private Room', 'Shared room': 'Shared Room'})
```

```python
df.rename(columns={'neighborhood': 'Neighborhood', 'room_type': 'Room_Type', 'accommodates': 'Accommodates',
                  'bedrooms': 'Bedrooms', 'number_of_reviews': 'Number_of_Reviews', 'wifi': 'Wifi', 'cable_tv': 'Cable_TV',
                  'washer': 'Washer', 'kitchen': 'Kitchen', 'price': 'Price(US_Dollars)'}, inplace=True)
```

```python
df.head()
```

```python
#Create a csv.
df.to_csv('air_final.csv', index=False)
```
